```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import nltk
```

```python
df = pd.read_csv("emoji_df.csv")
df.head(15)
```

Basic EDA

```python
print("length of df:",len(df))
print()
print("groups:",df['group'].unique())
print()
print("groups:",df['sub_group'].unique())
```

```python
# plot group frequencies
group_counts = df['group'].value_counts()
group_counts_sorted = group_counts.sort_values()
plt.barh(group_counts_sorted.index, group_counts_sorted.values)
plt.xlabel('Frequency')
plt.ylabel('Group')
plt.title('Frequency of Groups')
plt.show()
```
